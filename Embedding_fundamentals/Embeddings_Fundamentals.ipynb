```python
repo_id = "google-bert/bert-base-uncased"
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="google-bert/bert-base-uncased"
)

vector = embeddings.embed_query("Hello world")
```

```python
sentance = [
      "AI models learn patterns from data",
    "Machine learning systems detect trends automatically",
    "Algorithms analyze information to improve decisions",
    "A painter mixes colors on a canvas",
    "The train arrived late due to rain",
    "Regular exercise improves physical health",
    "Daily workouts strengthen the body",
    "Walking helps maintain overall wellness",
    "Stock markets fluctuate every day",
    "The laptop battery is fully charged"
]


vectors = embeddings.embed_documents(sentance)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(vectors)

df = pd.DataFrame(
    similarity_matrix,
    columns=[f"S{i+1}" for i in range(len(vectors))],
    index=[f"S{i+1}" for i in range(len(vectors))]
)

df = df.round(2)
print(df)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

model = SentenceTransformer("all-MiniLM-L6-v2")

sentences = [
    "AI models learn patterns from data",
    "Machine learning systems detect trends automatically",
    "Algorithms analyze information to improve decisions",
    "A painter mixes colors on a canvas",
    "The train arrived late due to rain",
    "Regular exercise improves physical health",
    "Daily workouts strengthen the body",
    "Walking helps maintain overall wellness",
    "Stock markets fluctuate every day",
    "The laptop battery is fully charged"
]

query = "AI systems learn from data"

sentence_embeddings = model.encode(sentences)
query_embedding = model.encode(query)

scores = cosine_similarity([query_embedding], sentence_embeddings)[0]

ranked = sorted(
    zip(sentences, scores),
    key=lambda x: x[1],
    reverse=True
)

for sent, score in ranked:
    print(f"{score:.3f}  |  {sent}")
```
